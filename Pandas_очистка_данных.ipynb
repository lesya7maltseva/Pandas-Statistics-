```python
import pandas as pd
import numpy as np

#считываем файл в датафрэйм, из описания файла выбираем нужные столбцы и с помощью iloc сохраняем в новый фрэйм, с которым и работаем
df_raw = pd.read_csv('raw_horses.csv', na_values= '?',usecols= [0,1,3,4,5,6,10,22],
                     names = ["surgery", "Age", "rectal_temperature", "pulse", "respiratory_rate", "temperature_of_extremities", "pain", "outcome"])
df_raw.describe()
```

Задание № 2

```python
df_work_nanfree = df_raw.dropna()
df_work_nanfree.describe(include = 'all')
```

```python
#квартили и межквартильный размах.определение выбросов
q1_rt = df_raw['rectal_temperature'].quantile(0.25)
q3_rt = df_raw['rectal_temperature'].quantile(0.75)
iqr_rt = q3_rt - q1_rt
lower_rt = q1_rt - (1.5 * iqr_rt)
higher_rt = q3_rt + (1.5 * iqr_rt)
print(iqr_rt)
print(lower_rt)
print(higher_rt)
```

```python
df_out_rt = df_raw[~df_raw['rectal_temperature'].between(lower_rt,higher_rt,inclusive = True)]
df_out_rt.dropna()
```

```python
q1_pulse = df_raw['pulse'].quantile(0.25)
q3_pulse = df_raw['pulse'].quantile(0.75)
iqr_pulse = q3_pulse - q1_pulse
lower_pulse = q1_pulse - (1.5 * iqr_pulse)
higher_pulse = q3_pulse + (1.5 * iqr_pulse)
print(iqr_pulse)
print(lower_pulse)
print(higher_pulse)
```

```python
# Все выбросы:
df_out_pulse = df_raw[~df_raw['pulse'].between(lower_pulse,higher_pulse,inclusive = True)]
df_out_pulse.dropna()
```

```python
q1_rr = df_raw['respiratory_rate'].quantile(0.25)
q3_rr = df_raw['respiratory_rate'].quantile(0.75)
iqr_rr = q3_rr - q1_rr
lower_rr = q1_rr - (1.5 * iqr_rr)
higher_rr = q3_rr + (1.5 * iqr_rr)
print(iqr_rr)
print(lower_rr)
print(higher_rr)
```

```python
# Все выбросы:
df_out_rr = df_raw[~df_raw['respiratory_rate'].between(lower_rr,higher_rr,inclusive = True)]
df_out_rr.dropna()
```

Задание № 3

```python
(df_raw.isna().mean() * 100).round(2)
```

```python
df_work = df_raw.copy()
# Поскольку в описании данных Возраста есть только 1 или 2, а в наших данных 9, логично предположить ошибку и заменить 9 на 2
df_work.loc[df_work['Age'] == 9,'Age'] = 2
```

```python
# Делаем предположение, что показатели пульса и частоты дыхания могут зависеть от возрастных групп:
df_1 = df_raw.loc[df_raw['Age'] == 1]
df_1.pulse.median()
```

```python
df_work.loc[(df_work['Age'] == 1) & (df_work['pulse'].isna()),'pulse'] = df_1['pulse'].median()
```

```python
df_2 = df_raw.loc[df_raw['Age'] == 2]
df_2['pulse'].median()
# Разница существенна, заполняем на основании возрастов
```

```python
df_work.loc[(df_work['Age'] == 2) & (df_work['pulse'].isna()),'pulse'] = df_2['pulse'].median()
```

```python
# для частоты дыхания как меняется медиана, в заисимости от возрастных групп и если брать общую:
print(df_1['respiratory_rate'].median())
print(df_2['respiratory_rate'].median())
print(df_raw['respiratory_rate'].median())
# изменения довольно существенные, стоит заменить пропуски медианой в зависимости от возраста
```

```python
df_work['respiratory_rate'] = df_work['respiratory_rate'].fillna(df_work.groupby('Age')['respiratory_rate'].transform('median'))
df_work
```

```python
# для частоты температуры как меняется медиана, в заисимости от возрастных групп и если брать общую:
print(df_1['rectal_temperature'].median())
print(df_2['rectal_temperature'].median())
print(df_raw['rectal_temperature'].median())
# изменения незначительные
# а если в зависимости от проведенной операции:
df_surgery = df_raw.loc[df_raw['surgery'] == 1]
df_therapy = df_raw.loc[df_raw['surgery'] == 2]
print(df_surgery['rectal_temperature'].median())
print(df_therapy['rectal_temperature'].median())
print(df_raw['rectal_temperature'].median())
# разница не существенна, предлагаю заполнить медианой по столбцу температуры без категорий
```

```python
df_work.loc[df_work['rectal_temperature'].isna(),'rectal_temperature'] = df_work['rectal_temperature'].median()
df_work
```

```python
# Предлагаю заполнение уровня боли сделать на основании категории проведения операции, поскольку такая взаимосвязь более логична, нежели возраста.
df_work['pain'] = df_work['pain'].fillna(df_work.groupby('surgery')['pain'].transform(lambda x: x.mode()[0]))
df_work
```

```python
# Оставшийся показатель является субъективной оценкой периферического кровообращения, из чего можно предполагать отсутствие его связи с остальными метриками,
# но проверим моду в зависимости от возраста и операций:
print(df_1['temperature_of_extremities'].mode())
print(df_2['temperature_of_extremities'].mode())
print(df_raw['temperature_of_extremities'].mode())

print(df_surgery['temperature_of_extremities'].mode())
print(df_therapy['temperature_of_extremities'].mode())
print(df_raw['temperature_of_extremities'].mode())
# Зависимость не подтвердилась:
```
