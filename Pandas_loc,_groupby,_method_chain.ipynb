Создайте столбец language, который проводит соответствие между языком и language_id по правилу:

А теперь давайте посчитаем - в скольких задачах в среднем один пользователь отправлял на выполнение код в каждой когорте по каждому языку.

Примечание: Задача не может быть одновременно и на SQL, и на Python.

Результатом должна стать «широкая» таблица, где:


строки - когорты, но только после января 2022 года (включительно)

столбцы - языки

значения - число задач, по которым пользователи в данной когорте отправляют код на выполнение по данному языку в среднем

```python
coderun = pd.read_csv('itresume-coderun.csv')
```

```python
dct_language = {3:'Python', 2:'SQL' }
```

```python
coderun['language'] = coderun.language_id.map(dct_language)
```

```python
coderun['cohort'] = pd.to_datetime(coderun['created_at']).dt.strftime('%Y-%m')
coderun
```

```python
coderun = coderun.loc[coderun['cohort']>='2022-01']
df = coderun.groupby(['cohort','language','user_id']).agg({'problem_id':'nunique'}).reset_index()
df = df.groupby(['cohort','language']).agg({'problem_id': 'mean'}).reset_index()
df.pivot_table(values= 'problem_id', index='cohort', columns='language')
```

через метод method chain (сначала кажется сложновато, но нереально круто выглядит и не надо городит огород как в примере сверху)

```python
df = (
    coderun.loc[coderun['cohort']>='2022-01']
    .groupby(['cohort','language','user_id'])
    .agg({'problem_id':'nunique'})
    .groupby(['cohort','language'])
    .agg({'problem_id': 'mean'})
    .pivot_table(values= 'problem_id', index='cohort', columns='language')
    )

df
```

Загрузите файлы в датафреймы Pandas и назовите их coderun и users. Далее создайте новый датафрейм df, который получается в результате:

Создайте новый датафрейм df2, который будет хранить информацию о том, сколько дней прошло между последним выполнением кода конкретного юзера и днем его регистрации. Столбец с разницей должен называться diff.
Округлите diff до ближайшего меньшего числа, кратного 50. Результат должен остаться целым числом.
Создайте датафрейм df3, в который запишите количество юзеров, сгруппированных по разнице дней.
Результат отсортируйте по убыванию количества user_id

```python
coderun = pd.read_csv('itresume-coderun.csv')
users = pd.read_csv('itresume-users-pandas.csv', sep = ';', skiprows=1)
```

```python
df = coderun.merge(users, left_on= 'user_id', right_on= 'id', how = 'left', suffixes=['_run', '_user'])
df
```

```python
df2 = (
    df.groupby(['user_id'])
    .apply(lambda x: pd.Series({'diff': max(pd.to_datetime(x['created_at']) - pd.to_datetime(x['date_joined'])).days}), include_groups=False)
    .reset_index()
    )

df2['diff'] = (df2['diff']/50).apply(np.floor)*50
```

```python
df2
```

```python
df3 = (
    df2.groupby('diff', dropna= False)
    .agg({'user_id':'count'})
    .reset_index()
    .sort_values('user_id', ascending= False)
    )
df3
```
