```python
import pandas as pd
```

```python
columns_movies = ["movie id", "movie title", "release date", "video release date", "IMDb URL", "unknown",
                  "Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary",
                  "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance",
                  "Sci-Fi", "Thriller", "War", "Western"]
columns_ratings = ["user id", "item id", "rating", "timestamp"]
```

```python
#1
df_movies = pd.read_csv("u.item.csv", sep = '|', encoding="windows_1258", names = columns_movies, header = None)
df_ratings = pd.read_csv("u.data.csv", sep = '\t', names = columns_ratings, header= None)
```

```python
df_movies.head()
```

```python
df_ratings.head()
```

```python
#2
df_ratings.groupby('user id')['rating'].count().sort_values(ascending=False)
```

```python
#3
df_best = df_ratings.loc[df_ratings['user id'] == 405]
df_best.rename(columns = {'rating':'ratings'},inplace=True)
df_best
```

```python
df_best = df_best.iloc[:,0:3]
df_best.head()
```

```python
#4
df_merged = df_movies.merge(df_ratings, how = 'left', left_on= 'movie id', right_on = 'item id')
df_merged = df_merged.groupby(['movie id','movie title']).agg({'rating':'sum','timestamp':'count'})
df_merged.head()
```

```python
df_merged.reset_index(inplace=True)
df_merged.rename(columns={'rating':'sum_rate','timestamp':'total_amount'}, inplace=True)

```

```python
df_best = df_best.merge(df_merged, how = 'left',left_on = 'item id', right_on = 'movie id')
df_best
```

```python
# проверяем пропуски. пропусков нет
df_best.describe()
```

```python
df_best = df_best.merge(df_movies,how = 'left',on = 'movie id')
```

```python
df_best
```

```python
# преобразуем к формату дата-время
df_best['release date'] = pd.to_datetime(df_best['release date'])
# вытаскиваем год
df_best.loc[:,'release date'] = df_best.loc[:,'release date'].dt.year
```

```python
df_best.drop(['movie id', 'movie title_y','video release date','IMDb URL','user id', 'item id'], axis = 1,inplace= True)
```

```python
df_best.head()
```

```python
df_best.columns
```

```python
#5
from sklearn.model_selection import train_test_split
```

```python
X = df_best[['sum_rate', 'total_amount', 'release date',
       'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy',
       'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror',
       'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War',
       'Western']]
y = df_best['ratings']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#6
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn import metrics
y_pred = lr.predict(X_test)
```

```python
#7
MSE = metrics.mean_absolute_error(y_test,y_pred)
MSE
```

```python
import findspark
findspark.init()
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").getOrCreate()
sc = spark.sparkContext
```

```python
#8
df_ratings2 = spark.read.csv('u.data.csv',inferSchema=True, sep='\t', header = False)
```

```python
#добавляем заголовки
df_ratings2 = df_ratings2.toDF("user_id", "item_id", "rating", "timestamp")
df_ratings2.show()
```

```python
df_movies2 = spark.read.csv('u.item.csv', sep= '|', inferSchema=True, header = False)
```

```python
df_movies2 = df_movies2.toDF("movie_id", "movie_title", "release date", "video release date", "IMDb URL", "unknown",
                  "Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary",
                  "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance",
                  "Sci-Fi", "Thriller", "War", "Western")
```

```python
df_movies2.show()
```

```python
df_merged2 = df_movies2.join(df_ratings2, df_movies2.movie_id == df_ratings2.item_id, how = 'left')
df_merged2.show()
```

```python
#9
df_avg_rating = df_merged2.groupBy('movie_title').avg('rating')
df_avg_rating.orderBy(['avg(rating)'], ascending = [False]).show()
```

```python
#10
genre_list = ["unknown", "Action", "Adventure", "Animation",
               "Children's", "Comedy", "Crime", "Documentary",
               "Drama", "Fantasy", "Film-Noir", "Horror",
               "Musical", "Mystery", "Romance","Sci-Fi",
               "Thriller", "War", "Western"]
genre_dict = {}
average_rating = {}
```

```python
df_temp = df_merged2.filter(df_merged2['Action'] == 1)
df_temp.agg({'rating':'avg'}).show()
```

```python
for genre in genre_list:
  df_temp = df_merged2.filter(df_merged2[genre] == 1)
  genre_dict[genre] = [data[0] for data in df_temp.select('rating').collect()]
```

```python
for genre in genre_dict.keys():
  average_rating[genre] = sum(genre_dict[genre])/len(genre_dict[genre])
```

```python
df_genre_rating = pd.DataFrame.from_dict(average_rating,orient='index')
```
